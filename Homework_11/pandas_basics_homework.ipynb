### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
ratings_stats = ratings.merge(movies, on='movieId', how='left')
```

```python
ratings_5_0 = ratings_stats[(ratings_stats['rating']==5.0)]
movieId_best = ratings_5_0['movieId'].value_counts().index[0]
print(movies[(movies['movieId']==movieId_best)].values[0][1])
```
