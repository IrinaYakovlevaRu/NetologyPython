# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
documents = [ {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
              {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
              {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}]
````

````python
directories = {'1': ['2207 876234', '11-2'],
               '2': ['10006'],
               '3': []}
````

````python
def get_doc_number():
    doc_number = input('Введите номер документа: ')
    return doc_number
````

````python
def get_dir_number():
    dir_number = input('Введите номер полки: ')
    return dir_number
````

````python
def get_doc_directory(doc_number):
    doc_directory = [key for key, value in directories.items() if doc_number in value]
    return doc_directory
````

````python
def get_doc_directories():
    doc_directories = ", ".join(list(directories.keys()))
    return doc_directories
````

````python
def function_p(doc_number):
    doc_owner = [document['name'] for document in documents if list(document.values())[1] == doc_number]
    if doc_owner:
        print(f'Владелец документа: {doc_owner[0]}') 
    else:
        print('Документ не найден в базе')
````

````python
def function_s(doc_number):
    doc_directory = get_doc_directory(doc_number)
    if doc_directory:
        print(f'Документ хранится на полке: {doc_directory[0]}') 
    else:
        print('Документ не найден в базе')
````

````python
def function_l():
    for document in documents:
        print(f'№: {document["number"]}, тип: {document["type"]}, владелец: {document["name"]}, полка хранения: {get_doc_directory(document["number"])[0]}')
````

````python
def function_ads(dir_number):
    if dir_number in directories:
        print(f'Такая полка уже существует. Текущий перечень полок: {get_doc_directories()}')
    else:
        directories[dir_number] = []
        print(f'Полка добавлена. Текущий перечень полок: {get_doc_directories()}')
````

````python
def function_ds(dir_number):
    if dir_number in directories:
        if directories[dir_number] == []:
            del(directories[dir_number])
            print(f'Полка удалена. Текущий перечень полок: {get_doc_directories()}')
        else:
            print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {get_doc_directories()}')
    else:
        print(f'Такой полки не существует. Текущий перечень полок: {get_doc_directories()}')
````

````python
def function_ad():
    new_doc_number = input('Введите номер документа: ')
    new_doc_type = input('Введите тип документа: ')
    new_doc_name = input('Введите владельца документа: ')
    new_doc_dir = input('Введите полку для хранения: ')
    if new_doc_dir in directories:
        documents.append({'type': new_doc_type, 'number': new_doc_number, 'name': new_doc_name})
        directories[new_doc_dir].append(new_doc_number)
        print('Документ добавлен. Текущий список документов:')
        function_l()
    else:
        print('Такой полки не существует. Добавьте полку командой ads') 
        print('Текущий список документов: ')
        function_l()
````

````python
def function_d(doc_number):
    document_d = [document for document in documents if list(document.values())[1] == doc_number]
    if document_d:
        documents.remove(document_d[0])
        directories[get_doc_directory(doc_number)[0]].remove(doc_number)
        print('Документ удален.')   
        print('Текущий список документов:')
        function_l()
    else:
        print('Документ не найден в базе') 
        print('Текущий список документов:')
        function_l()
````

````python
def function_m(doc_number,dir_number):
    if dir_number in directories:
        doc_directory = get_doc_directory(doc_number)
        if doc_directory:
            directories[doc_directory[0]].remove(doc_number)
            directories[dir_number].append(doc_number)
            print('Документ перемещен.')
            print('Текущий список документов: ')
            function_l()
        else:
            print('Документ не найден в базе.') 
            print('Текущий список документов: ')
            function_l()
    else:
        print(f'Такой полки не существует. Текущий перечень полок: {get_doc_directories()}')
````

````python
def main():
    while True:
        user_input = input('Введите команду: ')
        if user_input == 'p':
            function_p(get_doc_number())
        elif user_input == 's':
            function_s(get_doc_number())
        elif user_input == 'l':
            function_l()
        elif user_input == 'ads': 
            function_ads(get_dir_number()) 
        elif user_input == 'ds':
            function_ds(get_dir_number())
        elif user_input == 'ad':
            function_ad()
        elif user_input == 'd':
            function_d(get_doc_number())
        elif user_input == 'm':
            function_m(get_doc_number(),get_dir_number())    
        elif user_input == 'q':
            break
        else: 
            print('Команда не существует')
````

````python
main()
````
